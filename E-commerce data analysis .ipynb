```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px 
data=pd.read_csv("Sample - Superstore.csv",encoding='latin1')
data
```

<H3 style='color:orange;'>DATA CLEANING</H3>

```python
data.dropna(inplace=True)
```

<p style='color:violet;font-weight:bold;'>Remove the rows which are having null values.</p>

```python
data.info()
```

<p style='color:violet;font-weight:bold;'>The columns ['Order Date'] and ['Ship Date'] are in the object data type, so we need to convert them into the date format.<P>

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

<p style='color:violet;font-weight:bold;'>Remove the unwanted columns like ['Row ID'].</p>

```python
data.drop('Row ID',axis=1)
```

```python
data.describe()
```

```python
data.shape
```

<p style='color:violet;font-weight:bold;'>Adding a new column for 'Order Year'.</p>

```python
data['Order Year']=data['Order Date'].dt.year
```

```python
data
```

<H3 style='color:orange;'>VISUALIZING TRENDS</H3>

```python
yearly_sales=data.groupby('Order Year')['Sales'].sum().reset_index()
yearly_sales
```

<p style='color:violet;font-weight:bold;'>Obtaining Yearly sales</p>

```python
yearly_sales.plot(kind='bar',x='Order Year',y='Sales',color='black')
plt.gcf().set_facecolor('grey') 
plt.xlabel('YEAR',color='black')
plt.ylabel('SALES',color='black')
plt.title('YEARLY SALES TREND',color='black')
plt.show()
```

<p style='color:violet;font-weight:bold;'>The above bar graph represents the yearly trends of sales from year 2014 to 2017.</p>
<p style='color:violet;font-weight:bold;'>It suggests that sales are increasing every year exponentially.</p>
<p style='color:violet;font-weight:bold;'>Year 2017 has the highest sales of over 700k.</p>


```python
sales_by_region=data.groupby('Region')['Sales'].sum().reset_index()
sales_by_region
```

<p style='color:violet;font-weight:bold;'>Obtaining sales by region.</p>

```python
sales_by_region.plot(kind='pie',x='Region',y='Sales',labels=sales_by_region['Region'],autopct='%1.1f%%',wedgeprops={'width':0.3,'edgecolor':'white'},colors=['yellow','orange','purple','aqua'])
plt.title('Sales by Region',color='BLACK')
plt.gcf().set_facecolor('grey')
plt.ylabel('')
plt.legend(loc='lower left',title='REGION')
plt.show()
```

<p style='color:violet;font-weight:bold;'>This Pie chart shows the sales income produced by each region of USA.</p>
<p style='color:violet;font-weight:bold;'>The western region of USA contributing dominant sales of 31.6%.</p>

```python
avg_profit_per_customer=data['Profit'].mean()
print(avg_profit_per_customer)
```

<p style='color:violet;font-weight:bold;'>Average profit gained on a customer is 28.65689630778467.</p>

```python
data['Shipping time']=(data['Ship Date']-data['Order Date']).dt.days
data
```

```python
ship_mode_time=data.groupby('Ship Mode')['Shipping time'].mean().reset_index()
ship_mode_time.sort_values(by='Shipping time',inplace=True)
ship_mode_time
```

```python
plt.stem(ship_mode_time['Ship Mode'],ship_mode_time['Shipping time'],linefmt='black',markerfmt='teal',basefmt=' ')
plt.xlabel('Ship Mode')
plt.ylabel('Average Shipping Time(in days)')
plt.title('Average shipping Time by Ship Mode')
plt.gcf().set_facecolor('gray')
plt.show()
```

<p style='color:violet;font-weight:bold;'>The lollipop chart reveals significant variation in shipping efficiency across modes.</p>
<P style='color:violet;font-weight:bold;'> Average Shipping Times:<br>Same Day ~0 days,<br>First Class ~1–2 days,<br>Second Class ~3–4 days,<br>Standard Class ~6–7 days.</P>

```python
data['Order month']=data['Order Date'].dt.month
data
```

```python
monthly_sales=data.groupby(['Order Year','Order month'])['Sales'].sum().reset_index()
monthly_sales
```

```python
monthly_sales_in_2017=monthly_sales[monthly_sales['Order Year']==2017]
monthly_sales_in_2017.reset_index()
```

<p style='color:violet;font-weight:bold;'>Monthly analysis of sales in the year(2017).</p>

```python
monthly_sales_in_2017.plot(kind='line',x='Order month',y='Sales',color='g')
plt.title('MONTHLY SALES TREND IN PREVIOUS YEAR (2017)')
plt.gcf().set_facecolor('grey')
plt.xlabel('MONTH')
plt.ylabel('SALES')
plt.grid(axis='y')
plt.show()
```

<p style='color:violet;font-weight:bold;'>Monthly sales trend in the year 2017.</p>

```python
sales_profit_by_segment=data.groupby('Segment').aggregate({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment
```

```python
sales_profit_by_segment.plot(kind='bar',x='Segment',y=['Sales','Profit'],color=['black','teal'])
plt.grid(axis='y')
plt.ylabel('REVENUE (in millions)')
plt.xlabel('SEGMENT')
plt.title('SALES VS PROFIT IN EACH SEGMENT')
plt.gcf().set_facecolor('gray')
plt.show()
```

<p style='color:violet;font-weight:bold;'>The above plot represents total-sales and total-profit relation in each customer segment.</p>

```python
profit_by_category=data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

<P style='color:violet;font-weight:bold;'>Obtaining profits in each category.</p>

```python
profit_by_category.plot(kind='pie',y='Profit',labels=profit_by_category['Category'],autopct='%1.1f%%',wedgeprops={'width':0.3,'edgecolor':'black'},colors=['violet','yellow','aqua'])
plt.title('Profit by Category')
plt.ylabel('')
plt.gcf().set_facecolor('darkgrey')
plt.show()
```

<P style='color:violet;font-weight:bold;'>The above chart shows the profit gained in each category of products.</P>

```python
sales_by_state=data.groupby('State')['Sales'].sum().reset_index()
sales_by_state
```

```python
sales_by_state.loc[:, 'State Codes'] = sales_by_state['State'].map(state_codes)
sales_by_state
```

```python
fig = px.choropleth(
    sales_by_state,
    locations='State Codes',
    locationmode='USA-states',
    color='Sales',
    scope='usa',
    color_continuous_scale='reds',
    title='SALES BY STATE IN USA'
)
fig.show()
```

<p style='color:violet;font-weight:bold;'>The above Heat map represents the sales in each state of USA.</p>
<P style='color:violet;font-weight:bold;'>California and New York lead with sales exceeding 300,000, highlighting major market hubs.</P>
